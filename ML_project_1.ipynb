```python
# 1. Reset df_clean fresh from your original dataframe
df_clean = df.copy()

# 2. Map text categories to numbers safely
df_clean['Gender'] = df_clean['Gender'].map({'male': 0, 'female': 1})
df_clean['Smoking_Status'] = df_clean['Smoking_Status'].map({'yes': 0, 'no': 1})

# 3. Create Region dummies
geo_dummies = pd.get_dummies(df_clean['Geo_Region'], drop_first=True)
df_clean = pd.concat([df_clean.drop(columns=['Geo_Region']), geo_dummies], axis=1)

# 4. Create BMI categories and their dummies
df_clean['bmi_category'] = pd.cut(df_clean['Body_Mass_Index'],
                                  bins=[0, 18.5, 24.9, 29.9, float('inf')],
                                  labels=['underweight', 'normal', 'overweight', 'obese'])

df_clean = pd.get_dummies(df_clean, columns=['bmi_category'], drop_first=True)

# 5. Convert everything cleanly to integers
df_clean = df_clean.astype(int)
```

```python
df_clean
```

```python
from sklearn.preprocessing import StandardScaler
scale= StandardScaler()
col=['Age_Years','Body_Mass_Index','Children']
df_clean[col]=scale.fit_transform(df_clean[col])
```

```python
df_clean.head()
```

```python
import pandas as pd
from scipy.stats import pearsonr

selected_features = [
    'Age_Years', 'Gender', 'Body_Mass_Index', 'Children', 'Smoking_Status',
    'northwest', 'southeast', 'southwest',
    'bmi_category_normal', 'bmi_category_overweight', 'bmi_category_obese']

correlations = {}
for feature in selected_features:
    # pearsonr returns (coefficient, p_value). [0] grabs just the coefficient.
    score = pearsonr(df_clean[feature], df_clean['Cost'])[0]
    correlations[feature] = score

correlation_results = pd.DataFrame(correlations.items(), columns=['Feature', 'Correlation']).sort_values(by='Correlation', ascending=False)
print(correlation_results)
```

```python
import pandas as pd
from scipy.stats import chi2_contingency

cat_features = ['Gender', 'Smoking_Status']

df_clean['charges_bin'] = pd.qcut(df_clean['Cost'], q=4, labels=False)

alpha = 0.05
chi2_results = {}


for col in cat_features:

    contingency = pd.crosstab(df_clean[col], df_clean['charges_bin'])
    chi2_stat, p_val, _, _ = chi2_contingency(contingency)

    decision = 'Reject Null (Keep Feature)' if p_val < alpha else 'Fail to Reject Null (Drop Feature)'
    chi2_results[col] = {
        'chi2_statistic': chi2_stat,
        'p_value': p_val,
        'Decision': decision
    }


chi2_df = pd.DataFrame(chi2_results).T
print(chi2_df)
```
